## Data Visualisation
[31/01/2022]

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("../dataset/cs_langs.csv")
df
```

```python
# Barplot for minutes used
df.plot("Language","Minutes Used",kind="bar",figsize=(7,5))
plt.title("Language vs Minutes Used")
```

```python
# Using matplotlib
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_xlabel("Language")
ax.set_ylabel("Minutes Used")
ax.set_title("Programming Language Usage")
ax.bar(df.Language,df["Minutes Used"])
```

```python
# Horizontal Bar Graph
plt.barh(df.Language,df["Minutes Used"])
```

```python
# Different color for bar plot
plt.bar(df.Language,df["Minutes Used"],color=['red','blue','green'])
```

```python
# Error Bar : Understand how spread out the data is
# Longer the bar more is the variability(more spread out the data is)
lang_df = df.iloc[:,2:5]
bars = lang_df.mean(axis=1)
error_bars = lang_df.std(axis=1)
plt.bar(df.Language,df["Minutes Used"],yerr=error_bars)
```

## Renter Fraction
(% of people in that place who are rented)

```python
df = pd.read_csv("../dataset/renter_fraction.csv",sep="\t")
df.head()
```

```python
plt.plot(df["Year"],df["San Francisco"])
#plt.scatter(df["Year"],df["San Francisco"],color="red")
plt.xlabel("Year")
plt.ylabel("Renter Fraction(SF)")
plt.ylim(0,50)
```

```python
# Multiple plots
plt.plot(df["Year"],df["San Francisco"],label="SF",color="red")
plt.plot(df["Year"],df["USA"],label="USA",color="blue")
plt.plot(df["Year"],df["California"],label="CA",color="green")
plt.legend()
#plt.scatter(df["Year"],df["San Francisco"],color="red")
plt.xlabel("Year")
plt.ylabel("Renter Fraction(SF)")
plt.ylim(0,50)
```

```python
# avoid multiple plots
plt.xlabel("Year")
plt.ylabel("Renter Fraction(SF)")
plt.ylim(0,50)
[plt.plot(df["Year"],df[i], label=i) for i in df.columns[1:]]
```
